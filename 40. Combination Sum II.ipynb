Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target.

Each number in candidates may only be used once in the combination.

Note: The solution set must not contain duplicate combinations.

 

Example 1:

Input: candidates = [10,1,2,7,6,1,5], target = 8
Output: 
[
[1,1,6],
[1,2,5],
[1,7],
[2,6]
]
Example 2:

Input: candidates = [2,5,2,1,2], target = 5
Output: 
[
[1,2,2],
[5]
]

```python
class Solution(object):
    def combinationSum2(candidates, target):
        #Sort the candidates
        candidates.sort()
        n = len(candidates)
        ans = []

        #backtrack function
        def backtrack(idx, curr_sum, elem):
            if curr_sum == target:
                ans.append(elem[:])
                print("added answer")
                return
            
            if idx == n or curr_sum > target:
                print("return nothing")
                return
            
            elem.append(candidates[idx])
            print(f"index is {idx}")
            print(elem)
            backtrack(idx + 1, curr_sum + candidates[idx], elem)
            print("we pop")
            elem.pop()
            # skipping the duplicates
            while idx + 1 < n and candidates[idx] == candidates[idx+1]:
                print("we skip duplicates")
                idx += 1
            
            print(f"we call again at {idx + 1}")
            backtrack(idx + 1, curr_sum, elem)
            
        backtrack(0, 0, [])
        return ans
```

```python
candidates = [10,1,2,7,6,1,5]
target = 8
Solution.combinationSum2(candidates,target)
```
